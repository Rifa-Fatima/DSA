## Circular singly linked list

```python
class Node:
    def __init__(self,value=None):
        self.value = value
        self.next = None
    
class CircularSinglyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def __iter__(self):
        node=self.head
        while node:
            yield node
            if node.next==self.head:
                break
            node=node.next
            

# to  create CSLL of one node
    def createCSLL(self,value):
        node = Node(value)
        node.next = node
        self.head = node
        self.tail = node
        return "The CSLL has been created."

# to insert node in CSLL
    def insert(self,value,location):
        if self.head is None:
            return "The head reference is None."
        else:
            newNode = Node(value)
            if location == 0:
                newNode.next = self.head
                self.head = newNode
                self.tail.next = newNode
            elif location == -1:
                newNode.next = self.tail.next
                self.tail.next = newNode
                self.tail = newNode
            else:
                tempNode = self.head
                for _ in range(location-1):
                    tempNode = tempNode.next
                newNode.next = tempNode.next
                tempNode.next = newNode
            return "The insertion is successfully completed"

# to traverse in CSLL
    def traversalCSLL(self):
        if self.head is None:
            print("There is not any element for traversal")
        else:
            currentNode = self.head
            while currentNode is not None:
                print(currentNode.value)
                currentNode = currentNode.next
                if currentNode == self.tail.next:
                    break

# Searching in CSLL
    def searchCSLL(self,target):
        if self.head is None:
            return "There is no node "
        else:
            currentNode = self.head
            while currentNode:
                if currentNode.value == target:
                    return True
                currentNode = currentNode.next
                if currentNode == self.tail.next:
                    break

# to delete a node
    def deleteCSLL(self,location):
        if self.head is None:
            print("There is no element in the list")
        else:
            if location == 0:
                if self.head == self.tail:
                    self.head.next = None
                    self.head = None
                    self.tail = None
                else:
                    self.head = self.head.next
                    self.tail.next = self.head
            elif location == -1:
                if self.head == self.tail:
                    self.head.next = None
                    self.head = None
                    self.tail = None
                else:
                    node = self.head
                    while node:
                        if node.next == self.tail:
                            break
                        node = node.next
                    node.next = self.head
                    self.tail = node
                            
            else:
                tempNode = self.head
                for _ in range(location-1):
                    tempNode = tempNode.next
                tempNode.next = tempNode.next.next
                
# To delete entire CSLL
    def deleteEntireCSLL(self):
        self.head= None 
        self.tail.next= None
        self.tail = None              
```

```python
circularSLL = CircularSinglyLinkedList()
print(circularSLL.createCSLL(1))
circularSLL.insert(0,0)
circularSLL.insert(2,1)
circularSLL.insert(3,-1)
circularSLL.insert(4,2)
circularSLL.deleteCSLL(-1)
circularSLL.traversalCSLL()
print(circularSLL.searchCSLL(2))

print([node.value for node in circularSLL])
```
